### Importação das bibliotecas necessárias

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

### Criando uma carga de dados aleatória

```python
x = [ num for num in range(7) ]
y = [ 10, 15, 40, 30, 20, 25, 25]
z = [ 3, 5, 10, 7, 10, 12, 12]
```

### Plotando o gráfico de área

```python
# configura o estilo do gráfico
plt.style.use('dark_background')

# desempacotamento de tupla para figura e axes
fig, ax = plt.subplots(figsize=(24,8))

# plotagem do primeiro gráfico
ax.plot(x, y, 
        marker='o', 
        color='white', 
        markerfacecolor='white', 
        markersize=5,
        linewidth=1,
        linestyle='dotted')

# plotagem do segunda gráfico
ax.plot(x, z, 
        marker='o', 
        color='white',
        markerfacecolor='white', 
        markersize=5,
        linewidth=1,
        linestyle='dotted')

# preenchimento das áreas dos gráficos
ax.fill_between(x, y, color='turquoise')
ax.fill_between(x, z, color='limegreen')

# título do gráfico e rótulo dos eixos
ax.set_title('Gráfico de Área'.upper(), fontweight='bold', fontsize=28, pad=15)
ax.set_xlabel('Valores Eixo X', fontsize=16, labelpad=25)
ax.set_ylabel('Valores Eixo Y', fontsize=16, labelpad=25)

# altera o tamanho dos labels ticks dos eixos x e y
ax.set_xticklabels(x, fontsize=14)
ax.set_yticklabels([num for num in range(-5, 41, 5)], fontsize=14)

# tirando as bordas do gráfico
for border in ['top', 'right', 'bottom', 'left']:
  ax.spines[border].set_visible(False)

# aplicando o grid no eixo y e colocando-o atrás do gráfico
ax.set_axisbelow(True)
plt.grid(visible=True, color='grey', axis='y')

# apresentando o gráfico
plt.show()

# salva o gráfico em formato png
fig.savefig('grafico_area.png', dpi=300)
```
