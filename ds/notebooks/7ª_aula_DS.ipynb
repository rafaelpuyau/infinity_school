### Figure

A maneira mais fácil de criar uma nova Figura é com pyplot. Veja alguns exemplos:

* Figura vazia e sem axes:

    `fig = plt.figure()`

```python
fig = plt.figure()
```

* Figura com um axes:

    `fig, ax = plt.subplots()`

```python
fig, ax = plt.subplots()
```

* Figura com um _grid_ 2x2 de axes:

    `fig, axs = plt.subplots(2,2)`

```python
fig, axs = plt.subplots(2,2)
```

## Separando em 2 gráficos
- Como argumento do plt.subplots podemos passar o número de linhas e colunas que queremos no nosso plot
    - `plt.subplots(nrows= ,ncols= )`
- Nesse caso vamos precisar passar um índice para o ax (`ax[0], ax[1]`)

### Exemplo 01

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(12,12))
xpoints = np.array([0, 6])
ypoints = np.array([0, 250])

axs[0, 0].plot(xpoints, ypoints)
axs[0, 0].set_title('Gŕafico de Linha')
axs[0, 0].set_xlabel('Eixo X')
axs[0, 0].set_ylabel('Eixo Y')

axs[0, 1].set_title('Gŕafico em Branco')
axs[0, 1].set_xlabel('Eixo X')
axs[0, 1].set_ylabel('Eixo Y')

axs[1, 0].set_title('Gŕafico em Branco')
axs[1, 0].set_xlabel('Eixo X')
axs[1, 0].set_ylabel('Eixo Y')

axs[1, 1].scatter([1, 2, 3, 4], [1, 4, 2, 3])
axs[1, 1].set_title('Gráfico de Dispersão')
axs[1, 1].set_xlabel('Eixo X')
axs[1, 1].set_ylabel('Eixo Y')

plt.show()
```

## Melhorando a visualização
- Para aumentar o tamanho do gráfico podemos usar o `figsize=(x,y)`
- Se quisermos mudar a proporção dos gráficos, podemos usar o `gridspec_kw={'height_ratios': []}`

```python
# Cria uma figura contendo um único axes
fig, ax = plt.subplots(nrows=2,ncols=1,
                       gridspec_kw={'height_ratios': [2,1]}, figsize=(16, 12))

# "Plota" alguns dados no axes
from numpy.random import randint
import numpy as np

# "Plota" alguns dados no axes
from numpy.random import randint
x = [num for num in range(1, 9)]
z = [round(np.log10(num) * 10 , 2) for num in range(1, 9)]
y = [randint(1, 9) for _ in range(8)]

meses = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto']

# ax[0].plot(x, y, label='Números aleatórios', color='navy', linestyle='dotted', linewidth=2, marker='*')
ax[0].bar(x, y, label='Números aleatórios', color='plum', zorder=2)

ax[0].set_title('Gráfico de Exemplo 1', fontsize=18)
ax[0].set_xlabel('Meses', fontsize=14)
ax[0].set_xticks(x)
ax[0].set_xticklabels(meses)
plt.xticks(fontsize=14)
ax[0].set_ylabel('Quantidade', fontsize=14)

ax[0].grid(visible=True, axis='both')
# ax[0].legend(loc='upper right')
ax[0].legend(fontsize=12)

ax[1].plot(x, z, label='Logaritmo de 1 à 9', color='sienna', linestyle='dashed', linewidth=1, marker='o')

ax[1].set_title('Gráfico de Exemplo 2', fontsize=18)
ax[1].set_xlabel('Meses', fontsize=14)
ax[1].set_xticks(x)
ax[1].set_xticklabels(meses)
plt.xticks(fontsize=14)
ax[1].set_ylabel('Quantidade', fontsize=14)

ax[1].grid(visible=True, axis='both')
ax[1].legend(loc='upper left', fontsize=14)


plt.show()
```

## Vamos ajustar as barras para ficarem lado a lado
- Vamos definir a largura das barras
    - https://matplotlib.org/stable/plot_types/basic/bar.html
- Depois vamos deslocar o gráfico para os lados

```python
dados = {
    'X': [1,2,3,4,5,6],
    'Y': [120,110,130,145,118,125],
    'Z': [95,54,86,77,90,81]
}

base = pd.DataFrame(dados)
```

```python
# plot
fig, ax = plt.subplots(figsize=(16, 6))

ax.bar(base.X,base.Y, label="Y")
ax.bar(base.X,base.Z, label="Z")

# Colocando as barras lado à lado
# w=0.2
# ax.bar(base.X+w/2,base.Y, label="Y",width=w)
# ax.bar(base.X-w/2,base.Z, label="Z",width=w)

plt.grid(visible=True, axis='x')
plt.legend()
plt.show()
```

## Colocando rótulo nos dados
- Para isso, vamos usar o `annotate`
    - https://matplotlib.org/3.6.0/api/_as_gen/matplotlib.pyplot.annotate.html
    - Para usar o annotate, vamos passar `.annotate("<Texto>",<posicao(x,y)>)`
- Podemos posicionar o annotate exatamente em cima da barra usando a posição y
- Usando o `for`, é possível aplicar o annotate para todas as barras
- Por fim, vamos usar os parâmetros do annotate para melhorar a visualização
    - ha         -> alinhamento horizontal
    - xytext     -> deslocamento x,y do texto em relação a um referencial
    - textcoords -> referencial que vamos fazer o deslocamento acima
    - fontsize   -> tamanho da fonte
    - fontweight -> colocando em negrito

```python
# plot
fig, ax = plt.subplots(figsize=(16, 10))

ax.bar(base.X,base.Y, label="Y")
ax.bar(base.X,base.Z, label="Z")

# Colocando as barras lado à lado
# w=0.2
# ax.bar(base.X+w/2,base.Y, label="Y",width=w)
# ax.bar(base.X-w/2,base.Z, label="Z",width=w)

for i in range(0,6):   
    ax.annotate(
        'Diferença\n\n'+str(base.Y[i] - base.Z[i]),
        (base.X[i],base.Y[i]),
        ha="center",
        xytext=(0,-28),
        textcoords="offset points",
        fontweight='bold',
        fontsize=16
        )

plt.grid(visible=True, axis='x')
plt.legend()
plt.show()
```

```python
# plot
fig, ax = plt.subplots(figsize=(16,6))

w=0.3

ax.bar(base.X+w/2,base.Y, label="Y",width=w)
ax.bar(base.X-w/2,base.Z, label="Z",width=w)

for i in range(0,6):
    ax.annotate(
        base.Z[i],
        (base.X[i]-w/2,base.Z[i]),
        ha="center",
        xytext=(0,5),
        textcoords="offset points"
        )
    ax.annotate(
        base.Y[i],
        (base.X[i]+w/2,base.Y[i]),
        ha="center",
        xytext=(0,5),
        textcoords="offset points"
        )

plt.legend()
plt.show()
```

## Ajustando o eixo y para melhorar a visualização dos rótulos
- O `.set_yticks` vai permitir que alterar o intervalo do eixo y
- O `.yaxis.set_visible` permite que os valores do eixo y não fiquem visíveis

```python
# plot
fig, ax = plt.subplots(figsize=(16, 6))

w=0.3

ax.bar(base.X+w/2,base.Y, label="Y",width=w)
ax.bar(base.X-w/2,base.Z, label="Z",width=w)

for i in range(0,6):
    ax.annotate(
        base.Z[i],
        (base.X[i]-w/2,base.Z[i]),
        ha="center",
        xytext=(0,5),
        textcoords="offset points"
        )
    ax.annotate(
        base.Y[i],
        (base.X[i]+w/2,base.Y[i]),
        ha="center",
        xytext=(0,5),
        textcoords="offset points"
        )

ax.set_yticks([num for num in range(0, base['Y'].max() + 20, 10)])
ax.yaxis.set_visible(False)

plt.legend()
plt.show()
```

## Retirando bordas do gráfico
- Usando o `.spines[].set_visible(False)` é possível retirar as bordas de um gráfico

```python
# plot
fig, ax = plt.subplots(figsize=(16, 6))

w=0.3

ax.bar(base.X+w/2,base.Y, label="Peça feminina",width=w)
ax.bar(base.X-w/2,base.Z, label="Peça masculina",width=w)

for i in range(0,6):
    ax.annotate(
        base.Z[i],
        (base.X[i]-w/2,base.Z[i]),
        ha="center",
        xytext=(0,-20),
        textcoords="offset points",
        color='white',
        fontweight='bold',
        fontsize=16
        )
    ax.annotate(
        base.Y[i], 
        (base.X[i]+w/2,base.Y[i]),
        ha="center",
        xytext=(0,-20),
        textcoords="offset points",
        color='white',
        fontweight='bold',
        fontsize=16
        )


ax.set_xticks(base['X'])
ax.set_xticklabels(['Calças', 'Camisetas', 'Bermudas', 'Camisas', 'Tênis', 'Acessórios'])
# ax.set_yticks([0,170])
ax.yaxis.set_visible(False)

plt.xticks(fontsize=14)

for border in ['top', 'right', 'left']:
  ax.spines[border].set_visible(False)

plt.legend()
plt.show()
```

### Gerando valores aleatórios

```python
x = [ num for num in range(1, 13) ]
y = [ np.random.randint(10, 120) for _ in range(12)]
z = [ np.median(y).astype(int) for _ in range(12) ]
```

### Gerando o 1º gráfico

```python
'''Medidas do figsize em polegadas!

Para centímentros
cm = 1/2.54

Para pixels
px = 1/plt.rcParams['figure.dpi']
'''

# Centímetros
cm = 1/2.54

# Pixels
px = 1/plt.rcParams['figure.dpi']

fig, ax = plt.subplots(figsize=(16,6))

ax.bar(x, y, color='orange')
ax.plot(x, z, color='navy', marker='o', linestyle='dotted')

plt.show()
```

```python
'''Medidas do figsize em polegadas!

Para centímentros
cm = 1/2.54

Para pixels
px = 1/plt.rcParams['figure.dpi']
'''

# Centímetros
cm = 1/2.54

# Pixels
px = 1/plt.rcParams['figure.dpi']

fig, ax = plt.subplots(figsize=(16,6))

ax.bar(x, y, color='orange')
ax.plot(x, z, color='navy', marker='o', linestyle='dotted')

plt.title('Meu Gráfico de Exemplo 1', size=22, fontweight='bold')

ax.set_xticks([num for num in range(1,13)])
ax.set_xticklabels(
    ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez'],
    fontsize=16
)

plt.show()
```

```python
'''Medidas do figsize em polegadas!

Para centímentros
cm = 1/2.54

Para pixels
px = 1/plt.rcParams['figure.dpi']
'''

# Centímetros
cm = 1/2.54

# Pixels
px = 1/plt.rcParams['figure.dpi']

fig, ax = plt.subplots(figsize=(16,6))

ax.bar(x, y, color='orange')
ax.plot(x, z, color='navy', marker='o', linestyle='dotted')

for i in range(len(x)):
  ax.annotate(
      y[i],
      (i + 1, y[i]),
      ha='center',
      xytext=(0, 6),
      textcoords='offset points',
      fontsize=18,
      fontweight='bold',
      color='black' if y[i] >= np.mean(z) else 'red'
  )

plt.title('Meu Gráfico de Exemplo 1', size=22, fontweight='bold')

ax.set_xticks([num for num in range(1,13)])
ax.set_xticklabels(
    ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez'],
    fontsize=16
)
ax.set_yticks([num for num in range(0, max(y) + 30, 20)])
ax.set_yticklabels([num for num in range(0, max(y) + 30, 20)], fontsize=14)

plt.show()
```

```python
'''Medidas do figsize em polegadas!

Para centímentros
cm = 1/2.54

Para pixels
px = 1/plt.rcParams['figure.dpi']
'''

# Centímetros
cm = 1/2.54

# Pixels
px = 1/plt.rcParams['figure.dpi']

fig, ax = plt.subplots(figsize=(16,6))

ax.bar(x, y, color='orange')
ax.plot(x, z, color='navy', marker='o', linestyle='dotted')

ax.annotate(
    'Meta ' + str(np.mean(z)) + '%',
    (1, np.mean(z) + 5),
    fontsize=14,
    fontstyle='italic'
)

for i in range(len(x)):
  ax.annotate(
      y[i],
      (i + 1, y[i]),
      ha='center',
      xytext=(0, 16),
      textcoords='offset points',
      fontsize=18,
      fontweight='bold',
      color='black' if y[i] >= np.mean(z) else 'red'
  )

for side in ['top', 'left', 'right']:
  ax.spines[side].set_visible(False)

plt.title('Meu Gráfico de Exemplo 1', size=22, fontweight='bold')

ax.set_xticks([num for num in range(1,13)])
ax.set_xticklabels(
    ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez'],
    fontsize=16
)
ax.set_yticks([num for num in range(0, max(y) + 30, 20)])
ax.set_yticklabels([num for num in range(0, max(y) + 30, 20)], fontsize=14)

# Para colocar o grid atrás do gráfico
ax.set_axisbelow(True)

plt.grid(
    visible=True,
    axis='y',
    color='grey',
    linestyle='-',
    linewidth=0.3
)

plt.show()

# fig.savefig('meu_grafico_#1.png', dpi=300)
```
