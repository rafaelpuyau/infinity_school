## Importando as bibliotecas

```python
from matplotlib import pyplot as plt
import numpy as np
```

## Gerando valores aleatórios

```python
x = [ num for num in range(1, 13) ]
y = [ np.random.randint(10, 120) for _ in range(12)]
z = [ np.median(y).astype(int) for _ in range(12) ]
```

## Gerando o 1º gráfico

```python
'''Medidas do figsize em polegadas!

Para centímentros
cm = 1/2.54

Para pixels
px = 1/plt.rcParams['figure.dpi']
'''

# Centímetros
cm = 1/2.54

# Pixels
px = 1/plt.rcParams['figure.dpi']

fig, ax = plt.subplots(figsize=(16,6))

ax.bar(x, y, color='orange')
ax.plot(x, z, color='navy', marker='o', linestyle='dotted')

ax.annotate(
    'Meta ' + str(np.mean(z)) + '%',
    (1, np.mean(z) + 5),
    fontsize=14,
    fontstyle='italic'
)

for i in range(len(x)):
  ax.annotate(
      y[i],
      (i + 1, y[i]),
      ha='center',
      xytext=(0, 6),
      textcoords='offset points',
      fontsize=18,
      fontweight='bold',
      color='black' if y[i] >= np.mean(z) else 'red'
  )

for side in ['top', 'left', 'right']:
  ax.spines[side].set_visible(False)

plt.title('Meu Gráfico de Exemplo 1', size=22, fontweight='bold')

ax.set_xticks([num for num in range(1,13)])
ax.set_xticklabels(
    ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez'],
    fontsize=16
)
ax.set_yticks([num for num in range(0, max(y) + 30, 20)])
ax.set_yticklabels([num for num in range(0, max(y) + 30, 20)], fontsize=14)

# Para colocar o grid atrás do gráfico
ax.set_axisbelow(True)

plt.grid(
    visible=True,
    axis='y',
    color='grey',
    linestyle='-',
    linewidth=0.3,
)

plt.show()

fig.savefig('meu_grafico_#1.png', dpi=300)
```

## Gerando o 2º gráfico

```python
'''Medidas do figsize em polegadas!
Para centímentros
cm = 1/2.54

Para pixels
px = 1/plt.rcParams['figure.dpi']
'''

# Centímetros
cm = 1/2.54

# Pixels
px = 1/plt.rcParams['figure.dpi']

fig, ax = plt.subplots(figsize=(16,6))

colors = list(map(lambda valor: 'orange' if valor > np.mean(z) else 'red', y))

ax.bar(x, y, color=colors)
ax.plot(x, y, color='navy', marker='o', linestyle='dotted')

for i in range(len(x)):
  ax.annotate(
      y[i],
      (i + 1, y[i]),
      ha='center',
      xytext=(0, 6),
      textcoords='offset points',
      fontsize=16,
      fontweight='bold',
      color='black' if y[i] >= np.mean(z) else 'darkred'
  )

for side in ['top', 'left', 'right']:
  ax.spines[side].set_visible(False)

plt.title('Meu Gráfico de Exemplo 2', size=22, fontweight='bold')

ax.set_xticks([num for num in range(1,13)])
ax.set_xticklabels(
    ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez'],
    fontsize=16
)

plt.xticks(rotation=45)

ax.set_yticks([num for num in range(0, max(y) + 30, 20)])
ax.set_yticklabels([num for num in range(0, 121, 20)], fontsize=14)

# Para colocar o grid atrás do gráfico
ax.set_axisbelow(True)

plt.grid(
    visible=True,
    axis='both',
    color='grey',
    linestyle='-.',
    linewidth=0.3,
)
plt.show()

fig.savefig('meu_grafico_#2.png', dpi=300)
```
