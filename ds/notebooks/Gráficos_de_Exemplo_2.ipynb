## Importando as bibliotecas

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

## Gerando valores aleatórios

```python
x = [ num for num in range(3)]
y = [ np.random.randint(50, 100) for _ in range(3)]
```

## Plotagem do Gráfico

```python
plt.style.use('default')

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(24, 8))

# Gráfico 1
ax[0].bar(x[:-1], y[:-1], color='orange', edgecolor='black', zorder=2)
ax[0].bar(x[-1], y[-1], edgecolor='black', fill=False, zorder=2)

ax[0].set_title('Exemplo 1', fontweight='bold', fontsize=18)
ax[0].set_xticks(x)
ax[0].set_xticklabels(['Twitter', 'Instagram', 'TikTok'], fontsize=14)
ax[0].set_xlabel('Redes Sociais', fontsize=16)
ax[0].set_ylabel('Vissualizações', fontsize=16)
for side in ['top', 'left', 'right']:
  ax[0].spines[side].set_visible(False)
ax[0].spines['bottom'].set_linewidth(2.5)

# Gráfico 2
ax[1].bar(x[:-1], y[:-1], color='orange', zorder=2)
ax[1].bar(x[-1], y[-1], color='orange', alpha=0.3, zorder=2)

ax[1].set_title('Exemplo 2', fontweight='bold', fontsize=18)
ax[1].set_xticks(x)
ax[1].set_xticklabels(['Twitter', 'Instagram', 'TikTok'], fontsize=14)
ax[1].set_xlabel('Redes Sociais', fontsize=16)
ax[1].set_ylabel('Vissualizações', fontsize=16)
for side in ['top', 'left', 'right']:
  ax[1].spines[side].set_visible(False)
ax[1].spines['bottom'].set_linewidth(2.5)

plt.show()
```
